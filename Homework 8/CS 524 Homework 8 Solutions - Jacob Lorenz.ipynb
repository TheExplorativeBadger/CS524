# Problem 1

**Consider the problem:**

$$
    \min_{x_1,x_2} \frac{1}{2}(x_1^2 + x_2^2) \\
    \text{subject to:} \\
    1-x_1 \leq 0
$$

**a) Write down the solution of this problem and the optimal primal value p*.**

When looking at this problem, we notice a few things. First of all, we notice that there is no constraint on $x_2$. Second of all, we notice that the objective is a minimization and the objective scales directly proportional with the values of the variables $x_1$ and $x_2$, thus, it is in our best interest to minimize the variables' values as much as possible. Finally, we see that the feasible set is all $x_1$ such that $x_1 \geq 1$ Therefore, we see that the optimal solution will be found with a strict equality on the bound of $x_1$ and a value of 0 for $x_2$. Specifically, the solution to the problem is

$$
    x_1 = 1 \\
    x_2 = 0 \\
    \text{Objective Value} = p^* = \frac{1}{2}
$$

**b) Derive the Lagrangian dual function $g(\lambda)$ for $\lambda \in R$**

In the process of deriving the dual, it can be important to isolate the components to see how they map. The general standard form for the primal is as follows

$$
    \min_{x \in D} f_0(x) \\
    \text{subject to:} \\
    f_i(x) \leq 0 \;\;\;\; \forall i\\
    h_j(x) = 0 \;\;\;\; \forall j
$$

Furthermore, the Lagrangian dual has the folowing standard form

$$
    \max_{\lambda,v} g(\lambda,v) \\
    \text{subject to:} \\
    \lambda \geq 0
$$

where

$$
    g(\lambda,v) = \{\inf_{x \in D} L(x,\lambda,v)\} \leq p^*
$$

and

$$
    L(x,\lambda,v) = f_0(x) + \sum_{i=1}^m {\lambda_if_i(x)} + \sum_{j=1}^r {v_jh_j(x)}
$$

In the case of problem 1, we see there is only 1 inequality constraint and no equality constraints, and so we can map the problem as

$$
    f_0(x) = \frac{1}{2} (x_1^2 + x_2^2) \\
    f_1(x) = 1 - x_1 \\
    p^* = \frac{1}{2}
$$

Therefore, we can begin the derivation. We begin by noticing that there are no equality constraints and can thus define the simplified Lagrangian as

$$
    L(x,\lambda) = \frac{1}{2} (x_1^2 + x_2^2) + \lambda_1(1 - x_1)
$$

and 

$$
    g(\lambda) = \{inf_{x \in D} (\frac{1}{2} (x_1^2 + x_2^2) + \lambda_1(1 - x_1))\} \leq \frac{1}{2}
$$

By definition, the infimum of a subset $S$ of a partially ordered set $T$ is the greatest element in $T$ that is less than or equal to all elements of $S$, if such an element exists.

Now, if we take the partial derivative of the Lagrangian with respect to $x_1$ and $x_2$ and set to 0 to find the minimum, we get

$$
    \frac{\delta L}{\delta x_1} = x_1 - \lambda_1 = 0 => x_1 = \lambda_1 \\
    \frac{\delta L}{\delta x_2} = x_2 = 0
$$

Thus,

$$
    g(\lambda) = \frac{1}{2}\lambda_1^2 + \lambda_1 - \lambda_1^2 = \lambda_1 - \frac{1}{2}\lambda_1^2
$$

Finally, we now have everything we need to define the Lagrangian dual

$$
    \max_{\lambda} \lambda_1 - \frac{1}{2}\lambda_1^2\\
    \text{subject to:} \\
    \lambda_1 \geq 0
$$

**c) Find the solution of the Lagrangian dual problem $max_{\lambda \geq 0} \; g(\lambda)$ and write down the optimal dual objective d***

To find the solution of the Lagrangian dual derived in part b, we see that the optimum can be found with basic calculus. Namely, we will take the derivative of the Legrangian dual objective function, set it to 0, solve for $\lambda_1$ and then use this value to find the optimal dual objective $d^*$

$$
    \frac{\delta g(\lambda)}{\delta \lambda} = 1 - \lambda_1 = 0
$$

Solving this simple expression for $\lambda_1$, we get

$$
    \lambda_1^* = 1
$$

Hence

$$
    d^* = g(\lambda_1^*) = 1 - \frac{1}{2}(1)^2 = \frac{1}{2}
$$

**d) Is the Slater condition satisfied for this problem? Does strong duality hold, that is, $p^*$ = $d^*$?**

Slater's condition can be summarized as follows

$$
    \exists x_0 \in D : f_i(x_0) \lt 0 \;\;\;\; \forall i = 1,\dots,m
$$

In our case, there is only one constraint, namely

$$
    f_1(x) = 1-x_1 \leq 0
$$

Therefore, the feasible set is given by all ($x_1,x_2$) such that $x_1 \geq 1$. With this observation, we can see it is easy to find a ($x_1,x_2$) pair that satisfies the constraint with strict inequality, given by any ($x_1,x_2$) such that $x_1 \gt 1$. Hence, Slater's condition is indeed satisfied.

Strong duality does indeed hold, as seen from parts a) and c) where we determined that $p^* = d^* = \frac{1}{2}$.

# Problem 2
**Consider the problem:**

$$
    \min_{x_1,x_2} x_1*x_2 \\
    \text{subject to:} \\
    x_1^2 + x_2^2 - 1 \leq 0
$$

**a) Write down the solution of this problem and the optimal primal value p*.**

When we look at this problem, we notice a couple things. First of all, it might be easy to say we should just set both variables to 0; however, we quickly realize that there are no lower bounds on the variables, and that we can achieve a lower objective value by by having a negative and positive portion. So the next question, is what is the optimal balance that will fulfill the constraint and minimize the objective? One thing about math is there tends to be a lot of symmetry in problems, and that gave me the first hint on where to begin. Since we are dealing with decimals, we want both decimals to have the largest magnitude as possible, with one being the negative and one being the positive, so that the product between the two variables results in the lowest possible value.

To fulfill the symmetry, my approach was to treat $x_1$ and $x_2$ as the same value ignoring the sign. Thus, I converted the constraint to be 

$$
    2x^2 \leq 1 => x = +/- \sqrt{\frac{1}{2}} = +/- 0.70710678118
$$

From this, we can assign the negative or positive to either variable, and the opposite to the other variable. So, for example, we find the following solution to the problem:

$$
    x_1 = \sqrt{\frac{1}{2}} \\
    x_2 = -\sqrt{\frac{1}{2}} \\
    \text{Objective Value} = p^* = -\frac{1}{2}
$$

**b) Derive the Lagrangian dual function $g(\lambda)$ for $\lambda \in R$**

In the process of deriving the dual, it can be important to isolate the components to see how they map. The general standard form for the primal is as follows

$$
    \min_{x \in D} f_0(x) \\
    \text{subject to:} \\
    f_i(x) \leq 0 \;\;\;\; \forall i\\
    h_j(x) = 0 \;\;\;\; \forall j
$$

Furthermore, the Lagrangian dual has the folowing standard form

$$
    \max_{\lambda,v} g(\lambda,v) \\
    \text{subject to:} \\
    \lambda \geq 0
$$

where

$$
    g(\lambda,v) = \{\inf_{x \in D} L(x,\lambda,v)\} \leq p^*
$$

and

$$
    L(x,\lambda,v) = f_0(x) + \sum_{i=1}^m {\lambda_if_i(x)} + \sum_{j=1}^r {v_jh_j(x)}
$$

In the case of problem 2, we can map the features of the program to the appropriate Lagrange components

$$
    f_0(x) = x_1 * x_2 \\
    f_1(x) = x_1^2 + x_2^2 - 1 \\
    p^* = -\frac{1}{2}
$$

Thus, we can begin the derivation. We begin by noticing that there are no equality constraints and can thus define the simplified Lagrangian as

$$
    L(x,\lambda) = x_1 * x_2 + \lambda_1(x_1^2 + x_2^2 - 1)
$$

and hence,

$$
    g(\lambda) = \{inf_{x \in D} x_1 * x_2 + \lambda_1(x_1^2 + x_2^2 - 1)\} \leq -\frac{1}{2}
$$


We will now take the partial derivative of the Lagrangian with respect to $x_1$, $x_2$, and $\lambda$

$$
    \frac{\delta L}{\delta x_1} = x_2 + 2\lambda_1x_1 = 0 => x_1 = -\frac{x_2}{2\lambda} \\
    \frac{\delta L}{\delta x_2} = x_1 + 2\lambda_1x_2 = 0 => x_2 = -\frac{x_1}{2\lambda}\\
    \frac{\delta L}{\delta \lambda} = x_1^2 + x_2^2 - 1 = 0 => x_1 = +/- \sqrt{1-x_2^2} \;\; \text{and} \;\; x_2 = +/- \sqrt{1-x_1^2}
$$

Then, using the partial derivative of $x_1$, we have the following equivalent expression

$$
    \frac{\delta L}{\delta x_1} = \sqrt{1-x_1^2} + 2\lambda x_1 = 0
$$

Solving for $x_1$, we have

$$
    \sqrt{1-x_1^2} = -2\lambda x_1 \\
    1 - x_1^2 = 4\lambda^2x_1^2 \\
    1 = 4\lambda^2x_1^2 + x_1^2 = (4\lambda^2 + 1) x_1^2 \\
    => x_1 = +/- \sqrt{\frac{1}{4\lambda^2 + 1}}
$$

Thus, we can solve for $x_2$ with

$$
    x_2 = +/- \sqrt{1-x_1^2} = +/- \sqrt{1-\frac{1}{4\lambda^2 + 1}} = +/- \sqrt{\frac{4\lambda^2}{4\lambda^2 + 1}}
$$

Now that we have expressions for $x_1$ and $x_2$, we can use them in the original Lagrangian expression

$$
    L(x,\lambda) = +/- \sqrt{\frac{1}{4\lambda^2 + 1}} * +/- \sqrt{\frac{4\lambda^2}{4\lambda^2 + 1}} + \lambda (\frac{1}{4\lambda^2 + 1} + \frac{4\lambda^2}{4\lambda^2 + 1} -1) \\
    = +/- \sqrt{\frac{1}{4\lambda^2 + 1}} * +/- \sqrt{\frac{4\lambda^2}{4\lambda^2 + 1}}
$$

We can now redefine our Lagrangain dual function as

$$
    g(\lambda) = \{inf_{x \in D} (+/- \sqrt{\frac{1}{4\lambda^2 + 1}} * +/- \sqrt{\frac{4\lambda^2}{4\lambda^2 + 1}})\} \leq -\frac{1}{2}
$$


Since $\lambda \geq 0$, both of the terms in this expression will be positive, and so we need one of the expressions to take on its positive value, and the other expression to take on its negative value, it does not matter which. Simplifying $g(\lambda)$ by combining the square root terms, we have 

$$
    g(\lambda) = \{inf_{x \in D} (-\sqrt{\frac{1}{4\lambda^2 + 1} * \frac{4\lambda^2}{4\lambda^2 + 1}})\} \leq -\frac{1}{2}
$$

Therefore, we can finally define the Lagrangian Dual Program as

$$
    \max_{\lambda \geq 0} -\sqrt{\frac{4\lambda^2}{(4\lambda^2 + 1)^2}} \\
    \text{subject to:} \\
    \lambda \geq 0
$$

**c) Find the solution of the Lagrangian dual problem $max_{\lambda \geq 0} \; g(\lambda)$ and write down the optimal dual objective d***

To find the optimum of the Lagrangian dual, we start by taking the derivative of the dual's objective function and setting it to 0

$$
    \frac{\delta g(\lambda)}{\delta \lambda} = \frac{32\lambda^2}{(4\lambda^2+1)^2} - \frac{4}{4\lambda^2 + 1} = 0
$$

Simplfying and solving for $\lambda$, we get

$$
    \frac{32\lambda^2}{(4\lambda^2+1)^2} = \frac{4}{4\lambda^2 + 1} \\
    8\lambda^2 = 4\lambda^2 + 1 \\
    4\lambda^2 = 1 \\
    \lambda = +/- \sqrt{\frac{1}{4}} \\
    \lambda = +/- \frac{1}{2}
$$

So the solutions at $\lambda = \frac{1}{2}$ and $\lambda = -\frac{1}{2}$ are optimal for the expression; however, we have the constraint that $\lambda \geq 0$, and so we arrive at our final value of $\lambda^*$ given by

$$
    \lambda^* = \frac{1}{2}
$$

Using the value obtained for $\lambda^*$ in the objective function, we get

$$
    g(\lambda^*) = -\sqrt{\frac{4\frac{1}{2}^2}{(4\frac{1}{2}^2 + 1)^2}} \\
    = -\sqrt{\frac{1}{(1 + 1)^2}} = -\sqrt{\frac{1}{4}} \\
    = -\frac{1}{2}
$$

Therefore, we conclude that the solution to the Lagrangian dual for problem 2 is

$$
    \lambda^* = \frac{1}{2} \\
    d^* = -\frac{1}{2}
$$

One final check for sanity is to use this optimal value $\lambda^*$ to find the corresponding optimal $x_1$ and $x_2$.

We have already derived the following expressions

$$
    x_1 = +/- \sqrt{\frac{1}{4\lambda^2 + 1}} \\
    x_2 = -\frac{x_1}{2\lambda}
$$

Substituting $\lambda$ with our resulting value of $\lambda^* = \frac{1}{2}$ we get

$$
    x_1 = +/- \sqrt{\frac{1}{4\frac{1}{2}^2 + 1}} = +/- \sqrt{\frac{1}{2}}\\
$$

and

$$
    x_2 = -\frac{+/- \sqrt{\frac{1}{2}}}{2\frac{1}{2}} = - \; +/- \sqrt{\frac{1}{2}}
$$

This confirms our result from part a) where we found the magnitude of $x_1 = x_2 = \sqrt{\frac{1}{2}}$ and noted that one should be positive, one should be negative, and it does not matter which one is which. Symmetry is beautiful.

**d) Is the Slater condition satisfied for this problem? Does strong duality hold, that is, $p^*$ = $d^*$?**

Slater's condition can be summarized as follows
